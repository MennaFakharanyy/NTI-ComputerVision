```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import MobileNet
```

```python

def depthwise_conv(x, strides=1):
    x = layers.DepthwiseConv2D(
        kernel_size=3,
        strides=strides,
        padding="same",
        use_bias=False
    )(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    return x
```

```python

def pointwise_conv(x, filters):
    x = layers.Conv2D(
        filters,
        kernel_size=1,
        padding="same",
        use_bias=False
    )(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    return x

```

```python
def mobilenet_block(x, filters, strides=1):
    x = depthwise_conv(x, strides)
    x = pointwise_conv(x, filters)
    return x


```

```python
def MobileNet_Scratch(input_shape=(224,224,3), num_classes=1000):

    inputs = layers.Input(shape=input_shape)

    x = layers.Conv2D(
        32,
        kernel_size=3,
        strides=2,
        padding="same",
        use_bias=False
    )(inputs)

    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)

    # MobileNet Blocks
    x = mobilenet_block(x, 64)

    x = mobilenet_block(x, 128, strides=2)
    x = mobilenet_block(x, 128)

    x = mobilenet_block(x, 256, strides=2)
    x = mobilenet_block(x, 256)

    x = mobilenet_block(x, 512, strides=2)

    for _ in range(5):
        x = mobilenet_block(x, 512)

    x = mobilenet_block(x, 1024, strides=2)
    x = mobilenet_block(x, 1024)

    x = layers.GlobalAveragePooling2D()(x)

    outputs = layers.Dense(num_classes, activation="softmax")(x)

    model = models.Model(inputs, outputs)

    return model
```

```python
scratch_model = MobileNet_Scratch()

print("MobileNet From Scratch")
scratch_model.summary()

```

```python
base_model = MobileNet(
    weights="imagenet",
    include_top=False,
    input_shape=(224,224,3)
)

base_model.trainable = False

pretrained_model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(1000, activation="softmax")
])

print("\nPretrained MobileNet")
pretrained_model.summary()
```
